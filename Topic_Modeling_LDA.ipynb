```python
import nltk
import gensim
import pyLDAvis.gensim
import pandas as pd
from nltk.corpus import stopwords
from sklearn.datasets import fetch_20newsgroups

from gensim import corpora
import pyLDAvis
import pyLDAvis.gensim as gensimvis
import matplotlib.pyplot as plt
```

```python
nltk.download('stopwords')

# Step 2: Load the 20 Newsgroups Dataset
data = fetch_20newsgroups(subset='all')['data']

# Step 3: Preprocessing the Text
stop_words = set(stopwords.words('english'))
```

```python
def preprocess(text):
  tokens = gensim.utils.simple_preprocess(text, deacc = True)
  return [token for token in tokens if token not in stop_words]

processed_docs= [preprocess(doc) for doc in data]

dictionary = corpora.Dictionary(processed_docs)
dictionary.filter_extremes(no_below=15, no_above=0.5)  # Keep words that appear in at least 15 documents and less than 50% of documents
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]

num_topics = 10
```

```python
lda_model = gensim.models.LdaModel(corpus = corpus,
                                   id2word = dictionary,
                                   num_topics = num_topics,
                                   random_state = 42,
                                   update_every =1,
                                   passes = 10,
                                   alpha = 'auto',
                                   per_word_topics = True

                                   )
```

```python
topics = lda_model.print_topics(num_words = 5)
for idx, topic in topics:
  print(f"Topic {idx}: {topic}")
```
